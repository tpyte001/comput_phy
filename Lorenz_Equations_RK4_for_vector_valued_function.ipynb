Fourth Order Runge-Kutta Method (RK4)

$
x(t+h) = x(t) + \frac{1}{6}(k_1 + 2 k_2 + 2k_3 + k_4)
\\\\\\\ \text{where: } 
\\\\\\\ k_1 = hf(x,t)
\\\\\\\ k_1 = hf(x + \frac{k_1}{2}, t + \frac{h}{2})
\\\\\\\ k_2 = hf(x + \frac{k_2}{2}, t + \frac{h}{2})
\\\\\\\ k_3 = hf(x + k_3, t + h)
$

Generalizing to Multiple Dimensions:

$
\\ **r** = (x,y,z) = xi + yj + zk
\\ \frac{dr}{dt} = **f**(**r**,t) =
\\ \text{scalar form: }
\\ \frac{dx}{dt} f_x(x,y,z,t)
\\ \frac{dy}{dt} f_y(x,y,z,t)
\\ \frac{dz}{dt} f_z(x,y,z,t)
$

RK4:

$
r(t+h) = r(t) + \frac{1}{6}(k_1 + 2 k_2 + 2k_3 + k_4)
\\ \text{Vector Form: }
\\ ùê´(t+h) = ùê´(t) + \frac{1}{6}(**k**_1 + 2 **k**_2 + 2**k**_3 + **k**_4)
\\\\\\\ \text{where: } 
\\\\\\\ k_1 = hf(ùê´,t)
\\\\\\\ k_1 = hf(ùê´ + \frac{k_1}{2}, t + \frac{h}{2})
\\\\\\\ k_2 = hf(ùê´ + \frac{k_2}{2}, t + \frac{h}{2})
\\\\\\\ k_3 = hf(ùê´ + k_3, t + h)
$

r and k are vectors above, as is f

Solving Lorenz Equations

```python
import numpy as np
import matplotlib.pyplot as plt


a = 0
b = 50
bconst = 8/3
sigma = 10
rconst = 28
t = 0
N = 10000
h = (b-a)/N

def f(r,t):
  x = r[0]
  y = r[1]
  z = r[2]
  fx = sigma*(y-x)
  fy = 28*x-y-x*z
  fz = x*y - bconst * z
  fvec = np.array([fx,fy,fz],float)
  return fvec 

tlist = np.arange(a,b,h)
xlist = []
ylist = []
zlist = []
y = 1

rvec = [0,1,0]
for t in tlist:
  xlist.append(rvec[0])
  ylist.append(rvec[1])
  zlist.append(rvec[0])

  k1 = h * f(rvec,t)
  k2 = h * f((rvec + k1 / 2.0),(t + h / 2.0))
  k3 = h * f((rvec + k2 / 2.0),(t + h / 2.0))
  k4 = h * f((rvec + k3),(t + h))
  rvec = rvec + 1.0/6.0 * (k1 + 2*k2 + 2*k3 + k4)

plt.plot(xlist,ylist)
plt.plot()
```
