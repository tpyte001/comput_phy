```python
import numpy as np
import matplotlib.pyplot as plt

#this graph should be a wave

def dVoutdt(ylist,t):
  if np.floor(2*t) % 2 == 0:
    Vin = 1
  if np.floor(2*t) % 2 != 0:
    Vin = -1
  RC = 0.1
  return (Vin - ylist) / RC

a = 0
b = 10
N = 100
h = (b-a)/N

tlist = np.arange(a,b,h)
ylist = []
y = 1

for t in tlist:
  ylist.append(y)
  k1 = h * dVoutdt(y,t)
  k2 = h * dVoutdt((y + k1 / 2.0),(t + h / 2.0))
  k3 = h * dVoutdt((y + k2 / 2.0),(t + h / 2.0))
  k4 = h * dVoutdt((y + k3),(t + h))
  y = y + 1.0/6.0 * (k1 + 2*k2 + 2*k3 + k4)

graph2 = plt.plot(tlist,ylist)
plt.plot()
```
